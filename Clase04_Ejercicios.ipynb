## Ejercicio 1
<p>
Que, dadas dos edades, retorne la edad promedio si una de las edades es superior al doble de la otra, de lo contario retorne la diferencia de edades.
Utilice Listas.

</p>


```
#Solucion
import statistics as st
edades = []
edades.append(int(input('Ingresar edad 1: ')))
edades.append(int(input('Ingresar edad 2: ')))
if edades[0] > 2*edades[1]:
    promedio = st.mean(edades)
    print('El promedio es: {}'.format(promedio))
elif edades[1] > 2*edades[0]:
    promedio = st.mean(edades)
    print('El promedio es: {}'.format(promedio))
else:
    diferencia = abs(edades[0]-edades[1])
    print('La diferencia es: {}'.format(diferencia))
print('Fin')

```

Ingresando los dígitos de un numero de 3 cifras, devolver el cuadrado del número entero abc correspondiente si a+b es menor que c y de lo contrario devolver el número cba al cuadrado. 


```
#Solucion
valor = input('Ingresar valor: ')
a = valor[0]
b = valor[1]
c = valor[2]
print(a, b, c)
numero = int(valor)
centenas = numero // 100
decenas = numero//10 - centenas*10
unidades = numero - centenas*100 - decenas*10
print(centenas, decenas, unidades)
if centenas + decenas < unidades:
    resultado = numero ** 2
    texto = 'Se imprime el cuadrado'
else:
    resultado = unidades*100 + decenas * 10 + centenas
    texto = 'Se imprime cba'
print(texto, resultado)
print('Fin')
```

## Ejercicio 4
<p>
Elabore un algoritmo que lea tres triadas de coordenadas cartesianas en un espacio 3D (X, Y, Z) y determine el promedio de las distancias Euclidianas.
Validar que los datos se encuentren todos en el primer cuadrante para cada triada ordenada. Usar listas

$d_{E} = \sqrt{(x_{2}-x_{1})^{2}+(y_{2}-y_{1})^{2}+(z_{2}-z_{1})^{2}}$

![Imagen](https://docs.hektorprofe.net/cdn/wp-content/uploads/2015/10/cubo.png)
![Cuadrantes](https://concepto.de/wp-content/uploads/2019/09/plano-cartesiano-cuadrantes-e1569779047143.jpg)



```
#Solucion
#todas las x, y, z deben ser positivas
from math import sqrt, pow
from statistics import mean
vector1 = []
vector2 = []
vector3 = []

vector1.append(int(input('Ingresar coordenada x para el vector 1: ')))
vector1.append(int(input('Ingresar coordenada y para el vector 1: ')))
vector1.append(int(input('Ingresar coordenada z para el vector 1: ')))

vector2.append(int(input('Ingresar coordenada x para el vector 2: ')))
vector2.append(int(input('Ingresar coordenada y para el vector 2: ')))
vector2.append(int(input('Ingresar coordenada z para el vector 2: ')))

vector3.append(int(input('Ingresar coordenada x para el vector 3: ')))
vector3.append(int(input('Ingresar coordenada y para el vector 3: ')))
vector3.append(int(input('Ingresar coordenada z para el vector 3: ')))
#Validar que todos sean mayores o iguales a cero
if vector1[0] >= 0 and vector1[1] >= 0 and vector1[2] >= 0:
    v1 = True
if vector2[0] >= 0 and vector2[1] >= 0 and vector2[2] >= 0:
    v2 = True
if vector3[0] >= 0 and vector3[1] >= 0 and vector3[2] >= 0:
    v3 = True
distancia = []
if v1 and v2 and v3:
    #Puedo calcular las distancias
    distancia.append(sqrt((pow(vector1[0]-vector2[0], 2)+pow(vector1[1]-vector2[1], 2)+pow(vector1[2]-vector2[2], 2))))
    distancia.append(sqrt((pow(vector1[0]-vector3[0], 2)+pow(vector1[1]-vector3[1], 2)+pow(vector1[2]-vector3[2], 2))))
    distancia.append(sqrt((pow(vector2[0]-vector3[0], 2)+pow(vector2[1]-vector3[1], 2)+pow(vector2[2]-vector3[2], 2))))
    resultado = mean(distancia)
else:
    resultado = "Los valores no se encuentran en el primer cuadrante"
print(resultado)
```

## Ejercicio 5

Desarrolle un algoritmo para calcular el cobro del alquiler de videojuegos. El costo del alquiler es de $8.000 por cada juego. El cálculo del descuento se debe hacer con base en la cantidad de días de alquiler de acuerdo con la siguiente tabla: 

Cantidad días | Descuento
---|---
Hasta 3 | 0%
De 4 a 7 | 6%
De 8 a 10 | 9%
Mas de 10 | 15%

```
dias = 20 #@param {type:"slider", min:1, max:50, step:1}
juegos = 5 #@param {type:"slider", min:1, max:10, step:1}
#Solucion
```
